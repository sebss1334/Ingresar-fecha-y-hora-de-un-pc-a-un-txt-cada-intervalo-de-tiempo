```python
import pandas as pd
import matplotlib.pyplot as plt

# Carga el archivo CSV
df = pd.read_csv("recursos.csv", parse_dates=["Fecha"], dayfirst=True)

# Grafica los datos de memoria en funcin del tiempo
plt.figure(figsize=(10, 6))
plt.plot(df["Fecha"], df["Memoria (MB)"], marker='o', label="Memoria")
plt.xlabel("Fecha")
plt.ylabel("Memoria (MB)")
plt.title("Uso de memoria a lo largo del tiempo")
plt.legend()
plt.grid(True)
plt.show()
```
